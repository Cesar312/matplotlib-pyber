```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_file = "data/city_data.csv"
ride_file = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_file)
ride_df = pd.read_csv(ride_file)

# Combine the data into a single dataset
pyber_df = pd.merge(ride_df, city_df, on= "city", how= "inner")
# Display the data table for preview
pyber_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
pyber_grouped_df = pyber_df.groupby("city")
ride_cnt = pyber_grouped_df["ride_id"].count()
fare_avg = pyber_grouped_df["fare"].mean()
driver_cnt = pyber_grouped_df["driver_count"].mean()
city_type = pyber_grouped_df["type"].max()

pyber_summary_df = pd.DataFrame({"ride count": ride_cnt, "fare average": fare_avg, "driver count": driver_cnt, "type": city_type})

urban_summary_df = pyber_summary_df.loc[pyber_summary_df["type"] == "Urban", :]
suburban_summary_df = pyber_summary_df.loc[pyber_summary_df["type"] == "Suburban", :]
rural_summary_df = pyber_summary_df.loc[pyber_summary_df["type"] == "Rural", :]

# Build the scatter plots for each city types
ax1 = urban_summary_df.plot.scatter(x= "ride count", y= "fare average", marker= "o", color= "coral", edgecolors= "black",
            s= urban_summary_df["driver count"]*10, alpha= 0.7, label= "Urban")
ax2 = suburban_summary_df.plot.scatter(x= "ride count", y= "fare average", marker= "o", color= "skyblue", edgecolors= "black",
            s= suburban_summary_df["driver count"]*10, alpha= 0.7, label= "Suburban", ax = ax1)
ax3 = rural_summary_df.plot.scatter(x= "ride count", y= "fare average", marker= "o", color= "gold", edgecolors= "black",
            s= rural_summary_df["driver count"]*10, alpha= 0.7, label= "Rural", ax = ax1)

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fares ($)")
plt.grid()

# Create a legend
pyber_legend_1 = plt.legend(title= 'City Types', loc= "upper right", markerscale= .75, prop={'size': 9})
pyber_legend_1.legendHandles[0]._sizes = [30]
pyber_legend_1.legendHandles[1]._sizes = [30]
pyber_legend_1.legendHandles[2]._sizes = [30]
pyber_legend_1.get_title().set_fontsize('11') 

# Incorporate a text label regarding circle size
plt.text(42.5, 32.5, str(f"Note:\nCircle size corelates with driver count per city"))

# Save Figure
plt.savefig("Images/PyberRideSharingData.png") 
```

```python
# Show Figure
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_fare_amnt = urban_summary_df["fare"].sum().to_list()
suburban_fare_amnt = suburban_summary_df["fare"].sum().to_list()
rural_fare_amnt = rural_summary_df["fare"].sum().to_list()

# Build Pie Chart

labels = ["Urban", "Rural", "Suburan"] 
sizes = [urban_fare_amnt, rural_fare_amnt, suburban_fare_amnt]
colors = ["coral", "gold", "skyblue"]
explode = (0.1, 0, 0)
plt.pie(sizes, explode= explode, labels= labels, colors= colors, autopct= "%0.1f%%", shadow= True, startangle= 270)
plt.title("% of Total Fares by City Type") 

# Save Figure
plt.savefig("Images/TotalFaressByCity.png") 
```

```python
# Show Figure 
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_ride_cnt = urban_df["ride_id"].count().sum()
suburban_ride_cnt = suburban_df["ride_id"].count().sum()
rural_ride_cnt = rural_df["ride_id"].count().sum() 

# Build Pie Chart
labels = ["Urban", "Rural", "Suburan"] 
sizes = [urban_ride_cnt, rural_ride_cnt, suburban_ride_cnt]
colors = ["coral", "gold", "skyblue"]
explode = (0.1, 0, 0)
plt.pie(sizes, explode= explode, labels= labels, colors= colors, autopct= "%0.1f%%", shadow= True, startangle= 270)
plt.title("% of Total Rides by City Type") 

# Save Figure
plt.savefig("Images/TotalRidesByCity.png") 
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
urban_driver_df = city_df.loc[city_df["type"] == "Urban", :].groupby("type") 
suburban_driver_df = city_df.loc[city_df["type"] == "Suburban", :].groupby("type") 
rural_driver_df = city_df.loc[city_df["type"] == "Rural", :].groupby("type") 

urban_driver_cnt = sum(urban_driver_df["driver_count"].sum().to_list()) 
suburban_driver_cnt = sum(suburban_driver_df["driver_count"].sum().to_list()) 
rural_driver_cnt = sum(rural_driver_df["driver_count"].sum().to_list()) 

# Build Pie Charts
labels = ["Urban", "Rural", "Suburan"] 
sizes = [urban_driver_cnt, rural_driver_cnt, suburban_driver_cnt]
colors = ["coral", "gold", "skyblue"]
explode = (0.1, 0, 0)
plt.pie(sizes, explode= explode, labels= labels, colors= colors, autopct= "%0.1f%%", shadow= True, startangle= 270)
plt.title("% of Total Drivers by City Type") 

# Save Figure
plt.savefig("Images/TotalDriversByCity.png") 
```

```python
# Show Figure
plt.show() 
```
